```python
пример - ошибка линтера code-spell
'''Алгоритм решения проблем в компьютер саенс (OPEN SOURCE)'''
(проблемы: язык программирования, линтер, программы, CI/CD (deploy))
*опен сорс проект = можно прочесть код (код открыт). Его можно модифицировать и расширять.

*CI/CD (Continuous Integration / Continuous Deployment) — это практика разработки программного обеспечения, которая позволяет автоматизировать процесс интеграции и развертывания кода. 

1. ищем в чем ошибка (код ошибки, инструкция, в которой происходит ошибка) в файле. 

Нашли файл, в к-ром ошибка

2. Находим инструкцию, которая вызывает данную ошибку, код ошибки

3. Зайти в GitHub - ищем репозиторий по соответствующему линтеру, программе,апликейшну (app), в котором ошибка

как понять что это тот самый официальный репозиторий - много звезд от 1000 (если только он не новый)

4. Воспроизведение ошибки: проведение всех операций, которые способствуют ее появлению 

5. Ищем решение проблемы в ишьюсах репозитория через поиск

* пользователь заявляет о проблеме в Issues и также находит ответ. Но если не нашел, он заводит ишьюс (сообщение о прблеме), а также на основе него пул реквест, чтобы помочь решить самостоятельно (если способен). Помощь проекту - contribute opensorce (способствовать)

* каждый пул реквест прикреплен к какому то ишьюсу
название пул реквеста с номером ишьюса #<номер>

***pre-commit - общий линтер - это файл конфигурации, который запускает другие линтеры => не найдем ошибку определенного линтера. Туда будем 
в pre-commit записвать исправление ошибки

***Логирование ошибки - файл лога (полная запись ошибки)

???Какой код запускать в терминале, а какой прописывать в инструкции (pre-commit):
терминал: имя линтера - пробел - параметры линтера
инструкция: другое

копируем и ищем в .pre-commit-config.yaml
```

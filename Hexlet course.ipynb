#### git reset

нужно удалить только что сделанный по ошибке коммит из истории

Git позволяет удалять коммиты. Это опасная операция, которую нужно делать только в том случае, если речь идет про новые коммиты, которых нет ни у кого, кроме вас.
Пока коммит не отправлен в гитхаб (не сделали git push), его можно удалить. Когда уже есть во внешнем репозитории - нет


git reset --hard HEAD~

Флаг --hard означает полное удаление, не сохраняет изменения.


git reset HEAD~
Без --hard git reset отменит коммит, но не удалит его, а поместит все изменения этого коммита в рабочую директорию (но исключит из индекса), так что с ними можно будет продолжить работать.

(git reset --mixed HEAD~)
Без --hard команда работает в режиме --mixed, при котором изменения остаются в рабочей директории, но исключаются из индекса (unstage). Затем их можно исправить или отменить и выполнить новый коммит

Флаг HEAD~ означает «один коммит от последнего коммита». 
HEAD (переводится как «голова») — так обозначается последний сделанный коммит. 


git reset --hard HEAD~2
удалить два последних коммита




git reset --soft HEAD~1
Опция --soft отменяет коммит, но оставляет изменения в индексе (staging area).



### Изменение последнего коммита

Крайне часто разработчики делают коммит и сразу понимают, что забыли добавить часть файлов через git add.


git commit --amend
Если изменения еще не были отправлены во внешнюю систему, можно добавить изменения в текущий коммит. Для этого во время коммита добавляется флаг --amend

Пример:
echo 'experiment with amend' >> INFO.md
echo 'experiment with amend' >> README.md

git add INFO.md
git commit -m 'add content to INFO.md and README.md'
# Забыли сделать подготовку README.md к коммиту

git add README.md
git commit --amend


В реальности --amend не добавляет изменения в существующий коммит. Этот флаг приводит к откату коммита через git reset и выполнению нового коммита с новыми данными. Поэтому мы и видим ровно один коммит, хотя команда git commit выполнялась два раза (первый раз — когда сделали ошибочный коммит).

git commit --amend --no-edit
Чтобы не открывался редактор для ввода описания коммита к команде git commit --amend можно добавить опцию --no-edit. В этом случае описание коммита не изменится


git commit --amend -m <сообщение>
Если хотим добавить в последний коммит изменения и поменять сообщение (название) коммита

#### git branch

Переключение по коммитам отражается только на содержимом рабочей директории. Больше нигде не видно, где мы находимся. 
Из-за этого немало программистов забывают, где они находятся. 
Они начинают работать и очень удивляются, когда не получается выполнить коммит.

команда git branch - способ узнать место нахождения. 
когда мы находимся на последнем коммите, Git покажет такой вывод
(* main)

Но если прямо сейчас загружен коммит из прошлого, то вывод станет таким:
* (HEAD detached at e6f625c)

```python
### git checkout

Вы правы, git checkout в основном используется для переключения между ветками или для просмотра определенных коммитов. Однако, его также можно использовать для восстановления отдельных файлов или всего рабочего каталога до состояния, которое было в определенном коммите.


Восстановление отдельных файлов:

Чтобы восстановить файл до состояния, которое было в определенном коммите, используйте следующую команду:

git checkout <коммит> <путь/к/файлу>
⦁ <коммит>: Хэш коммита, тег или имя ветки, до состояния которой вы хотите восстановить файл. Можно использовать HEAD для последнего коммита, HEAD^ для предпоследнего и т.д.
⦁ <путь/к/файлу>: Путь к файлу, который вы хотите восстановить.

Пример:

git checkout HEAD~1 src/main.cpp
Эта команда восстановит файл src/main.cpp до состояния, которое было в предпоследнем коммите.



Восстановление всего рабочего каталога:

Чтобы восстановить весь рабочий каталог до состояния определенного коммита, используйте следующую команду (будьте осторожны, эта команда перезапишет все незакоммиченные изменения!):

git checkout <коммит> -- .


Важно понимать:

⦁ git checkout <коммит> без указания файлов: Эта команда переключит вас на указанный коммит, создавая "detached HEAD" состояние. Вы будете находиться в состоянии, как будто бы вы сделали checkout на временную анонимную ветку. Любые изменения, которые вы сделаете в этом состоянии, будут потеряны, если вы переключитесь на другую ветку, не создав новую ветку из этого состояния.

⦁ git checkout <коммит> <путь/к/файлу>: Эта команда не переключает вас на другой коммит. Она только изменяет указанный файл в вашем рабочем каталоге. Изменения нужно будет закоммитить, как обычно.

⦁ git checkout . (без указания коммита): Эта команда отменяет изменения в рабочем каталоге, возвращая файлы к состоянию последнего коммита. Будьте осторожны, эта команда также перезапишет все незакоммиченные изменения!

В контексте восстановления удаленного stash, git checkout может быть использован, если вы знаете хеш коммита, который был создан при применении stash. Найти этот хеш можно с помощью git reflog, если stash был удален недавно. Однако, это не самый надежный способ восстановления stash, и лучше использовать другие методы, если это возможно.
```

### Игнорирование файлов

В процессе работы над любым проектом в директории с кодом создаются файлы, которые не являются частью исходного кода. Все эти файлы можно условно разделить на несколько групп:

Инструментарий
Служебные файлы, добавляемые операционной системой — например, .DS_Store в MacOS
Конфигурационные и временные файлы редакторов — например, .idea или .vscode

Временные файлы
Логи — в них содержится полезная информация для отладки, которая собирается во время запуска и работы приложения
Кеши — файлы, которые нужны для ускорения разных процессов

Артефакты
Результаты сборки проекта — например, после компиляции или сборки фронтенда
Зависимости, которые устанавливаются во время разработки — например, node_modules или vendor
Результаты выполнения тестов — например, информация о покрытии кода тестами
Все это в обычной ситуации не должно попадать в репозиторий. В этом уроке вы узнаете, как проигнорировать эти файлы.

### Stash

Самая простая ситуация - ваши изменения не пересекаются с изменениями по срочной задаче => можете внести исправления, добавить их в индекс, закоммитить и запушить. 
А если изменения нужно делать в тех файлах, с которыми вы работаете прямо сейчас?
В Git существует набор команд, позволяющий прятать изменения в рабочей директории и восстанавливать их при необходимости. 

git stash
пример
echo 'fff' FILE.md
git add FILE.md
git status

(On branch main
Your branch is up to date with 'origin/main'.

Changes to be committed:
  (use "git restore --staged <file>..." to unstage)
    new file:   FILE.md)

# Прячем файлы
# После этой команды пропадут все измененные файлы
# независимо от того, добавлены они в индекс или нет
git stash

(Saved working directory and index state WIP on main: e7bb5e5 update README.md)

git status

(On branch main
Your branch is up to date with 'origin/main'.

nothing to commit, working tree clean) 
не отражается

Команда git stash не удаляет файлы. Они попадают в специальное место внутри директории .git на временное хранение. Эта команда не трогает новые файлы, так как они еще не являются частью репозитория:



git stash pop
чтобы вернуть спрятанные изменения после выполнения всех нужных изменений в чистой рабочей директории 
Файлы вернулись в том виде, в котором они попали в стэш (stash).

Можно много раз отправлять изменения стэш и вернуться они последовательно
Стэш в Git работает по принципу стека. Он позволяет сохранить внутрь любое количество изменений и восстановить их в обратном порядке:

git stash #1

# Изменяем файлы
git stash #2

# Возвращаем последние изменения
git stash pop #2

# Возвращаем предпоследние изменения
git stash pop #1

### Открытые проекты 

Проекты с открытым исходным кодом или OpenSource (Open Source Software, OSS) — это программное обеспечение с общедоступным кодом. Любой пользователь может увидеть код, а в большинстве случаев — даже повлиять на него.

Один из важнейших механизмов GitHub — запрос на включение изменений. Коротко его называют пулреквест (pull request). Именно этот механизм позволяет легко и непринужденно вливаться в разработку любых проектов


#### Клонирование
создаем копию репозитория в своем аккаунте - кнопкой Fork
Fork - Ответвление от основной кодовой базы другого репозитория на GitHub



#### Запрос на включение изменений кода
пулреквест - Запрос на внесение изменений из одной ветки в другую внутри репозитория или между форками
После того как мы залили изменения на GitHub, в его интерфейсе произойдут изменения. На странице склонированного репозитория появится кнопка Pull request:



Если ее нажать, то откроется страница, на которой можно указать название пулреквеста и его описание.

После отправки пулреквеста в оригинальном репозитории на странице Pull requests отобразится ваш запрос. Теперь остается ждать, когда разработчики библиотеки рассмотрят пулреквест. В итоге они:

Либо примут реквест
Либо отклонят его
Либо зададут уточняющие вопросы и предложат доработать реквест



#### Исправления прямо на GitHub
В более простых ситуациях, когда достаточно исправить текст или опечатку, GitHub позволяет сделать пулреквест прямо из своего интерфейса.

Для этого достаточно открыть любой файл проекта и нажать на иконку редактирования. После завершения редактирования GitHub сам предложит создать пулреквест с этим изменением.
